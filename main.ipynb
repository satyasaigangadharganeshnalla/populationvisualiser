```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from datetime import datetime as dt
import pycountry
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
paises = []

for country in pycountry.countries:
    paises.append(country.name)
print(paises)
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df["Country Name"].unique()
```

```python
df2 = df[df['Country Name'].isin(paises)]
df2.info()
```

```python
for col in df2.columns:
    print('_'*50)
    print(col)
    print(f'uniques: {df2[col].nunique()}')
```

```python
cols = df2.columns[1:65]
df2 = df2.groupby(['Country Name'])[cols].sum().stack().reset_index()
df2.columns=['country','year','population']
df2['year'] = df2['year'].str[0:4]
df2['year'] = pd.to_datetime(df2['year'], format='%Y')
df2.head(3)
```

```python
fig = px.line(df2.groupby('year')['population'].sum().reset_index(name="total_population"),
             x='year',
             y='total_population',
             title='Population Variation')
fig.show()

```

```python
px.line(df2.groupby(['country','year'])['population'].sum().reset_index(),
        x='year', 
        y='population', 
        color='country',
        title='Population growth',
        height=800,
        width=1200        
       )
```

```python
df3 = df2.groupby('year')['population'].sum().reset_index()
df3['population variation'] = df3['population'].diff().fillna(0)
px.line(df3, 
        x='year',
        y='population variation',
        title= 'population variation',
        height=600,
        width=900)
```

```python
df4 = df2.query("year=='2022-01-01'")
df4
```

```python
fig = px.choropleth(df4.groupby('country')['population'].sum().reset_index(), 
                    locations='country',
                    locationmode='country names',
                    color= 'population',
                    hover_name='country',
                    color_continuous_scale='Rainbow')
fig.update_geos(projection_type="natural earth", showcoastlines=True)
fig.update_layout(title_text='World Map 2022 - Population')
fig.show()
```
